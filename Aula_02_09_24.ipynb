# **Regressão Linear**

```python
#Equação Linear
# y = b0 + b1 X
#x = variavel preditora
#y = Oq quer prever
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
doenca_pulmao_csv = pd.read_csv('LungDisease.csv')
```

```python
doenca_pulmao_csv.plot.scatter(x="Exposure", y="PEFR")
plt.show()
```

```python
preditoras = ["Exposure"]
alvo = "PEFR"
```

```python
model = LinearRegression()
model.fit(doenca_pulmao_csv[preditoras], doenca_pulmao_csv[alvo])
print(f'Interceptação: {model.intercept_:.3f}')
print(f'Coeficiente de exposição: {model.coef_[0]:.3f}')
print(f'Equeção: Y = {model.intercept_:0.3f}+ + {model.coef_[0]:.3f}')
```

```python
fig, ax = plt.subplots()
## atribuição dos valores limites dos eixos x e y
ax.set_xlim(0, 23)
ax.set_ylim(295, 450)
## atribuição dos rótulos dos eixos x e y
ax.set_xlabel('Exposure')
ax.set_ylabel('PEFR')
## obtém os valores extremos da equação de regressão para desenhar os
## extremos da reta
ax.plot((0, 23), model.predict(pd.DataFrame({'Exposure': [0, 23]})))
## posiciona o intercepto (coeficiente b_0) na reta
ax.text(0.4, model.intercept_, r'$b_0$', size='larger')
plt.show()
```

```python
## computa os erros do modelo
previsto = model.predict(doenca_pulmao_csv[preditoras])
erros = doenca_pulmao_csv[alvo] - previsto

## visualização gráfica dos erros
## dados do dataset
ax = doenca_pulmao_csv.plot.scatter(x='Exposure', y='PEFR')
## dados previstos
ax.plot(doenca_pulmao_csv.Exposure, previsto)

for x, y_real, y_previsto in zip(doenca_pulmao_csv.Exposure,
                                 doenca_pulmao_csv.PEFR,
                                 previsto):
    ax.plot((x,x), (y_real, y_previsto), '--', color='C1')

plt.show()


#Ponto azul = valor real
#Linha azul = valor Previsto
#LInha laranja = erro
```
